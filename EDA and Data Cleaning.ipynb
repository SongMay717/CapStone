## cleaning up player positions

#### Create sub dataset

```python
subs = players[players['Club_Position']=='Sub']
```

```python
subs.head()
```

```python
subs['positions'] = subs['Preffered_Position']
```

```python
subs.head()
```

#### Create Res player dataset

```python
res = players[players['Club_Position'] == 'Res']
```

```python
res.head()
```

```python
res['positions'] = res['Preffered_Position']
```

```python
res.head()
```

#### Create non-sub, non-res player dataset

```python
non_sub_res = players[(players['Club_Position'] != 'Res') & (players['Club_Position'] != 'Sub')]
```

```python
non_sub_res.head()
```

```python
non_sub_res['positions'] = non_sub_res['Club_Position']
```

```python
non_sub_res.head()
```

```python
dataframes = [subs, res, non_sub_res]
```

```python
df = pd.concat(dataframes)
```

```python
df.head()
```

```python
df['positions'].value_counts()
```

```python
df = df.groupby("positions").filter(lambda x: len(x) > 181)
```

```python
df.head(2)
```

```python
df.shape
```

```python
df['positions'].value_counts()
```

```python
conditions = [
    (df['Rating'] > 84),
    (df['Rating'] < 85) & (df['Rating'] >70),
    (df['Rating'] < 71) & (df['Rating'] >65),
    (df['Rating'] < 64) & (df['Rating'] >49), 
    (df['Rating'] < 50)]
choices = [4, 3 ,2, 1, 0]

df['Category'] = np.select(conditions, choices)
```

```python
df['Category'].value_counts()
```

### Creating sub datasets according to player positions

```python
#Creating Goalkeeper position dataframe
df_gk = df[df['positions'] == 'GK']
```

```python
df_gk.head(2)
```

```python
df_gk.shape
```

```python
#Creating defense position dataset
df_def = df[(df['positions'] == 'CB')|
            (df['positions'] == 'LB')|
            (df['positions'] == 'RB')|
            (df['positions'] == 'RCB')|
            (df['positions'] == 'LCB')|
            (df['positions'] == 'CB/RB')|
            (df['positions'] == 'CB/LB')]
```

```python
df_def.head(2)
```

```python
df_def.shape
```

```python
#Creating mid-field position dataframe

df_mid = df[(df['positions'] == 'CM')|
            (df['positions'] == 'CAM')|
            (df['positions'] == 'CDM')|
            (df['positions'] == 'LCM')|
            (df['positions'] == 'RCM')|
            (df['positions'] == 'CDM/CM')|
            (df['positions'] == 'CM/CDM')|
            (df['positions'] == 'RDM')|
            (df['positions'] == 'LDM')]
```

```python
df_mid.head(2)
```

```python
df_mid.shape
```

```python
#Createing wing position dataframe

df_wing = df[(df['positions'] == 'RM')|
             (df['positions'] == 'LM')|
             (df['positions'] == 'RW')|
             (df['positions'] == 'LW')]
```

```python
df_wing.shape
```

```python
#Creating striker position dataframe
df_st = df[(df['positions'] == 'ST')|
           (df['positions'] == 'RS')|
           (df['positions'] == 'LS')]
```
